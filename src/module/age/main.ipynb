```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import widgets
```

```python
# Завантаження даних
doctors_df = pd.read_csv('doctors.csv')
patients_df = pd.read_csv('patients.csv')
operations_df = pd.read_csv('operations.csv')
appointments_1_df = pd.read_csv('appointments_1.csv')
appointments_2_df = pd.read_csv('appointments_2.csv')

appointments = pd.concat([appointments_1_df,appointments_2_df])

merged_data = pd.merge(appointments,doctors_df,on='doctor_id')
merged_data = pd.merge(merged_data,operations_df,on='operation_id')
merged_data = pd.merge(merged_data,patients_df,on='patient_id')
merged_data
```

```python
#a отримані за весь час
merged_data['total'] = merged_data['quantity'] * merged_data['price_per_unit']
merged_data
```

```python
def get_age_cateogory(birth):
   age = pd.to_datetime('now').year - pd.to_datetime(birth).year
   if age <= 16:
      return 'child'
   elif (age > 16) & (age <= 60):
      return 'adult'
   else:
        return 'senior'
        


merged_data['category'] = merged_data['date_of_birth'].apply(get_age_cateogory)
merged_data
```

```python
dropdown = widgets.Dropdown(
    options=merged_data['category'].unique(),
    value = merged_data['category'][2]
)
display(dropdown)
```

```python
choosen_category = dropdown.value
category_data = merged_data[merged_data['category'] == choosen_category]
patients_by_operations = category_data.groupby('last_name_patient')['quantity'].sum().reset_index()
category_data
```

```python

doctor_operations_table = merged_data.groupby(['last_name_doctor', 'operation_name'])['quantity'].sum().reset_index()
doctor_operations_table = doctor_operations_table.sort_values(by='quantity',ascending=False)
doctor_operations_table = doctor_operations_table.groupby('last_name_doctor').head(1)
doctor_operations_table
```
