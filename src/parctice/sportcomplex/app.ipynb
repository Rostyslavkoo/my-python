```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Visitor:
    def __init__(self, surname, id_visitor, money):
        self._surname = surname
        self._id_visitor = id_visitor
        self._some_of_money = money

    @property
    def surname(self):
        return self._surname

    @property
    def id_visitor(self):
        return self._id_visitor

    @property
    def some_of_money(self):
        return self._some_of_money
```

```python
class TrainingType:
    def __init__(self, name, id_number, cost_per_session):
        self._name = name
        self._id_number = id_number
        self._cost_per_session = cost_per_session

    @property
    def name(self):
        return self._name

    @property
    def id_number(self):
        return self._id_number

    @property
    def cost_per_session(self):
        return self._cost_per_session
```

```python
class TrainingInfo:
    def __init__(self, date, visitor_id, training_type_id):
        self._date = date
        self._visitor_id = visitor_id
        self._training_type_id = training_type_id

    @property
    def date(self):
        return self._date

    @property
    def visitor_id(self):
        return self._visitor_id

    @property
    def training_type_id(self):
        return self._training_type_id
```

```python
visitors = pd.read_csv('visitors.csv')
trainingType = pd.read_csv('training_type.csv')

merged_visitors_and_trainingType = pd.merge(visitors, trainingType, left_on='id_visitor', right_on='id_number')

print(merged_visitors_and_trainingType)


```
